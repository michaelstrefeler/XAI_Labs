- Make sure to be able to answer the questions bellow. You can write the answer / copy paste the definition here if it helps you: 

    - What is a linguistic variable ? It's a variable that has a set of **_linguistic values_** (e.g. "cold", "warm", "hot")
    - ... a membership function ? A membership function defines the boundaries of each linguistic value of a variable.
    - ... a Antecedent ? An antecendent is composed of a linguistic variable and its linguistic value (e.g. "cold" from linguistic variable "Temperature"). An antecendent can be set to NOT e.g. "Temperature" is NOT "cold". 
    - ... a rule ? A fuzzy rule is composed of one or more antecedents (the part(s) before the THEN), one or more consequents (the part(s) after the THEN), and a fuzzy operator such as AND or OR operators. IF antecedant(s) THEN consequent(s)
    - ... the aggregation function ? The aggregation function determines how each rule is combined
    - ... the defuzzification ? The step where you go from fuzzy logic to a numerical value
    - ... the difference between a Mamdani and a Singleton Fuzzy System ? Mamdani: In this type, both antecedents and consequents use fuzzy sets. Singleton: In this type, consequents are constant values (i.e. fuzzy sets that concentrate its membership values into a single point).
    - ... a default rule ? a fuzzy rule that has no antecedent (e.g. ELSE Tourists is LOW) which ensure that at least one fuzzy rule is active for any given input. The default rule is a fuzzy one, meaning that its activity is also a continuous value.

## Part two: implementation of the car acceleration problem

````python
%matplotlib ipympl

import numpy as np
import matplotlib.pyplot as plt

import numpy as np
import matplotlib.pyplot as plt

from lfa_toolbox.core.lv.linguistic_variable import LinguisticVariable
from lfa_toolbox.core.mf.lin_piece_wise_mf import LinPWMF
from lfa_toolbox.core.mf.trap_mf import TrapMF
from lfa_toolbox.core.mf.triangular_mf import TriangularMF
from lfa_toolbox.core.mf.singleton_mf import SingletonMF

from lfa_toolbox.core.rules.fuzzy_rule import FuzzyRule, Consequent, Antecedent
from lfa_toolbox.core.rules.default_fuzzy_rule import DefaultFuzzyRule
from lfa_toolbox.core.fis.fis import FIS, OR_max, AND_min, COA_func, MIN
from lfa_toolbox.core.fis.singleton_fis import SingletonFIS

from lfa_toolbox.view.lv_viewer import LinguisticVariableViewer
from lfa_toolbox.view.fis_viewer import FISViewer
from lfa_toolbox.view.fis_surface import show_surface
````


### A) Basic implementation (Mamdani model)

In this exercise we will look into the car acceleration problem that you have seen in the lectures and transform it, step by step, into an **interpretable fuzzy system**. 

**Tip: While you are doing the exercises, you can compare your results with the surface plots shown in the lecture notes.**

The car acceleration problem is a problem where you need to decide which **action** between "**push**", "**release**" or "**nothing**" you must perform knowing the **speed** and the **speed change** (acceleration) of your car to maintain the same speed.

We can model this problem with this initial proposition:


<img src="data:image/png;base64, 
" />

The figure above tells you which combination of inputs leads to which output. The goal here is to first implement a basic FIS and then to reduce its complexity to make it more interpretable.


````python
# LV car_speed
car_speed = LinguisticVariable(name='speed', ling_values_dict={
    'slow': LinPWMF([-0.2, 1], [0, 0]),
    'ok': LinPWMF([-0.2, 0], [0, 1], [0.15, 0]),
    'fast': LinPWMF([0, 0], [0.15, 1], [1, 1])
})

# LV speed change
car_acc = LinguisticVariable(name='speed_change', ling_values_dict={
    'slowing': LinPWMF([-0.3, 1], [0, 0]),
    'constant': LinPWMF([-0.3, 0], [0, 1], [0.3, 0]),
    'rising': LinPWMF([0, 0], [0.3, 1])
})

# LV action
car_action = LinguisticVariable(name='action', ling_values_dict={
    'release': LinPWMF([-1, 1], [0, 0]),
    'nothing': LinPWMF([-1, 0], [0, 1], [1, 0]),
    'push': LinPWMF([0, 0], [1, 1])
})

# Rules
car_rules = [
    FuzzyRule(ant_act_func=AND_min,
              ants=[Antecedent(car_speed, 'fast'),
                    Antecedent(car_acc, 'constant')],
              cons=[Consequent(car_action, 'release')], impl_func=MIN),

    FuzzyRule(ant_act_func=AND_min,
              ants=[Antecedent(car_speed, 'fast'),
                    Antecedent(car_acc, 'rising')],
              cons=[Consequent(car_action, 'release')], impl_func=MIN),

    FuzzyRule(ant_act_func=AND_min,
              ants=[Antecedent(car_speed, 'slow'),
                    Antecedent(car_acc, 'constant')],
              cons=[Consequent(car_action, 'push')], impl_func=MIN),

    FuzzyRule(ant_act_func=AND_min,
              ants=[Antecedent(car_speed, 'slow'),
                    Antecedent(car_acc, 'slowing')],
              cons=[Consequent(car_action, 'push')], impl_func=MIN),

    FuzzyRule(ant_act_func=AND_min,
              ants=[Antecedent(car_speed, 'fast'),
                    Antecedent(car_acc, 'slowing')],
              cons=[Consequent(car_action, 'nothing')], impl_func=MIN),

    FuzzyRule(ant_act_func=AND_min,
              ants=[Antecedent(car_speed, 'ok'),
                    Antecedent(car_acc, 'slowing')],
              cons=[Consequent(car_action, 'nothing')], impl_func=MIN),

    FuzzyRule(ant_act_func=AND_min,
              ants=[Antecedent(car_speed, 'ok'),
                    Antecedent(car_acc, 'constant')],
              cons=[Consequent(car_action, 'nothing')], impl_func=MIN),

    FuzzyRule(ant_act_func=AND_min,
              ants=[Antecedent(car_speed, 'ok'),
                    Antecedent(car_acc, 'rising')],
              cons=[Consequent(car_action, 'nothing')], impl_func=MIN),

    FuzzyRule(ant_act_func=AND_min,
              ants=[Antecedent(car_speed, 'slow'),
                    Antecedent(car_acc, 'rising')],
              cons=[Consequent(car_action, 'nothing')], impl_func=MIN)
]

# Fuzzy Systems
fis = FIS(rules=car_rules, aggr_func=np.max, defuzz_func=COA_func)

# Surface 
fig, ax = plt.subplots(figsize=(8, 8), subplot_kw=dict(projection='3d'))
show_surface(fis, ax=ax, title="Car Problem Mamdani", x_label="speed", y_label="speed_change", z_label="action",
                 n_pts=15, x_range=(-1, 1), y_range=(-1, 1), z_range=(-1, 1))
````

## Questions 

- (1 pt) Why this system is a Mamdani FIS ?

- (1 pt) What is showing the surface plot ? is it before or after defuzzification ?

*Réponse:*

This is a Mamdani FIS, because both antecedents and consequents use fuzzy sets (both use fuzzy/linguistic variables)

- *Why this system is a Mamdani FIS ?* 

This is a Mamdani FIS because both the antecedents and the consequents are expressed using fuzzy sets and linguistic variables. In particular, the input variables (speed and acceleration) are described by fuzzy linguistic terms such as "low", "medium", or "high", and the output (action) is also defined using fuzzy linguistic values like "push", "release", or "nothing". This is a hallmark of the Mamdani approach, which contrasts with other FIS types (such as Sugeno) where the output is often a numerical function of the inputs. Additionally, Mamdani systems typically require a defuzzification step to produce a crisp output, reinforcing the interpretability and human-readability of the system.

- *What is showing the surface plot ? is it before or after defuzzification ?* 

The surface plot shows how the crisp output of the FIS changes in response to different combinations of input values. It visualizes the behavior of the entire fuzzy inference system, including fuzzification, rule evaluation, aggregation, and defuzzification.
Therefore, the plot represents the system’s output after defuzzification - it shows the final, crisp action the system would take for each pair of input values.

- *Which rules are triggered and why? (which variables and membership functions lead to each rule’s activation?)*

The first rule states that if speed is fast and speed_change is not slowing, then the action is "release". At `speed = 1`, the input fully belongs to the fuzzy set "fast", as shown by the top-left membership plot where the activation is 1. The speed_change of 0.22 is a positive value, so it does not belong to the fuzzy set "slowing", which usually covers negative values. Therefore, the condition "not slowing" is fully satisfied as well, leading to strong activation of the first rule. The second rule, which would produce the action "push", is not triggered because the input speed = 1 does not belong to the fuzzy set "slow", making the overall rule activation zero. The default rule is not used here, since one of the defined rules is activated.
